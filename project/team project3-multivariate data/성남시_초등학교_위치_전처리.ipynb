```python
import pandas as pd
```

```python
sc = pd.read_csv('./data/전국초중등학교위치표준데이터.csv', encoding='cp949')
```

```python
el = sc[sc['학교급구분'] == '초등학교']
```

```python
el = el[['학교명', '소재지도로명주소', '위도', '경도']]
```

```python
el['지역'] = el['소재지도로명주소'].apply(lambda x: x.split()[0])
```

```python
el['시'] = el['소재지도로명주소'].apply(lambda x: x.split()[1])
```

```python
el['구'] = el['소재지도로명주소'].apply(lambda x: x.split()[2])
```

```python
el
```

```python
seong_df = el[(el['지역'] == '경기도') & (el['시'] == '성남시') & (el['구'] == '분당구')]
```

```python
seong_df = seong_df.drop(columns = {'지역', '시', '구'})
```

```python
seong_df
```

```python
seong_df.to_csv('분당구초등학교.csv', index=False)
```
