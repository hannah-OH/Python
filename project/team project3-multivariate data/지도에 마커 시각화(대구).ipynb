### 위경도 변경

```python
# https://developers.kakao.com/docs/latest/ko/local/dev-guide
# 카카오 REST API

def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK 키번호"} # 키번호는 개인정보라 삭제. 개인적으로 발급 후 사용 권장 
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    return api_json
```

```python
get_location("서울특별시 금천구 디지털로9길 23")
```

```python
# 위경도 변환 모델

def result_location(data, i):
    api_json = get_location(data['주소'][i])
    if api_json['documents']:
        address = api_json['documents'][0]['address']
        data.loc[i,'x'] = float(address['y'])       
        data.loc[i,'y'] = float(address['x']) 
    else:
        data.loc[i,'x'], data.loc[i,'y'] = None, None
```

```python
# 실행부 - 해법

i = 0
while i<=len(hb_dae['주소'])-1:
    try:
        result_location(hb_dae, i)
        i+=1
    except:
        result_location(hb_dae, i)
        i+=1
```

```python
# 실행부 - 미래엔

i = 0
while i<=len(mirae_dae['주소'])-1:
    try:
        result_location(mirae_dae, i)
        i+=1
    except:
        i+=1
```

### 행정구역 경계
* http://data.nsdi.go.kr/dataset/15144

```python
# 행정구역 데이터
geo_path = ('./geo/LARD_ADM_SECT_SGG_대구.json')
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
# 빈 지도 생성
dae_aed = folium.Map(location=[35.8714354, 128.601445], zoom_start=11)
```

```python
# 스마트해법 마커 추가 (빨간색)
for i in hb_dae.index:
    x = hb_dae.loc[i, 'x']
    y = hb_dae.loc[i, 'y']

    marker = folium.CircleMarker(location=[x, y],
                                  radius=5, line_color='red', fill_color='red', fill_opacity=1)
    marker.add_to(dae_aed)

# 미래엔수학 마커 추가 (파란색)
for i in mirae_dae.index:
    x = mirae_dae.loc[i, 'x']
    y = mirae_dae.loc[i, 'y']

    marker = folium.CircleMarker(location=[x, y],
                                  radius=5, line_color='blue', fill_color='blue', fill_opacity=1)
    marker.add_to(dae_aed)

# 초등학교 마커 추가 (노란색)
for i in el.index:
    x = el.loc[i, 'x']
    y = el.loc[i, 'y']

    marker = folium.CircleMarker(location=[x, y],
                                  radius=5, line_color='yellow', fill_color='yellow', fill_opacity=1)
    marker.add_to(dae_aed)

# 대구 정보 표기
folium.GeoJson(geo_str, name='json_data').add_to(dae_aed)

# 지도 저장 및 표시
dae_aed.save('./dae_aed.html')
dae_aed
```
