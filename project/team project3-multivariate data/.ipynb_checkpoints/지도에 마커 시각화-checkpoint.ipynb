### 위경도 변경

```python
# https://developers.kakao.com/docs/latest/ko/local/dev-guide
# 카카오 REST API

def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK 키번호"}
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    return api_json
```

```python
get_location("경기 용인시 기흥구 기흥역로58번길 56")
```

```python
# 위경도 변환 모델

def result_location(data, i):
    api_json = get_location(data['주소'][i])
    if api_json['documents']:
        address = api_json['documents'][0]['address']
        data.loc[i,'x'] = address['x']; 
        data.loc[i,'y'] = address['y']        
    else:
        data.loc[i,'x'], data.loc[i,'y'] = None, None
```

```python
# 실행부 - 해법

i = 0
while i<=len(hb['주소'])-1:
    try:
        result_location(hb, i)
        i+=1
    except:
        result_location(i)
        i+=1
```

```python
# 실행부 - 푸르넷

i = 0
while i<=len(pur['주소'])-1:
    try:
        result_location(pur, i)
        i+=1
    except:
        result_location(pur, i)
        i+=1
```

```python
# 실행부 - 미래엔

i = 0
while i<=len(mira['주소'])-1:
    try:
        result_location(mira, i)
        i+=1
    except:
        result_location(mira, i)
        i+=1
```
