```python
mu = pd.read_excel('./지역_번호.xlsx')
```

```python
num_list = []

for i in range(len(mu)):
    num_list.append(mu.loc[i, '번호'])
```

### 한번에 하고싶은데 잘안되서 고민중

```python
# 크롤링 시작 
mirae = [] # 결과물이 저장되는 리스트

for i in num_list:
    # url을 변경해가며 모든 지역 돌기
    
    ind = 1
    
    mirae_search_url = f'https://www.miraenmath.co.kr/search/neighborhood.html?areaCd1={i}'
    driver.get(mirae_search_url)
    
    while True:
        try:
            title = driver.find_element(by='xpath', value=f'//*[@id="class"]/tr[{ind}]/td[1]').text
            add = driver.find_element(by='xpath', value=f'//*[@id="class"]/tr[{ind}]/td[2]').text
            
            mirae.append([title, add])
            
            ind += 1
            
        except Exception as e:  
            #print(f"An error occurred: {e}")
            print(ind)
            print(i)
            continue  # 또는 continue 사용하여 계속 진행
```

```python
mirae = []  # 결과물이 저장되는 리스트

for i in num_list:
    ind = 1

    mirae_search_url = f'https://www.miraenmath.co.kr/search/neighborhood.html?areaCd1={i}'
    driver.get(mirae_search_url)

    while True:  # 무한 루프로 변경
        try:
            # XPath로 요소를 찾을 때 예외가 발생할 수 있으므로, 일단 요소를 찾고 없을 경우 예외가 발생
            title_element = driver.find_element(by='xpath', value=f'//*[@id="class"]/tr[{ind}]/td[1]')
            add_element = driver.find_element(by='xpath', value=f'//*[@id="class"]/tr[{ind}]/td[2]')
            
            title = title_element.text
            add = add_element.text
            
            mirae.append([title, add])
            
            ind += 1
            
        except NoSuchElementException:  # 찾는 요소가 없을 경우의 예외 처리
            break
        except Exception as e:  # 다른 예외 처리 (디버깅을 위해 모든 예외를 캐치)
            print(f"An error occurred: {e}")
            continue  # 또는 continue 사용하여 계속 진행

```
