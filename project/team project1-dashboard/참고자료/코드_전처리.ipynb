```python
# 2022년의 데이터만 추출, df2로 추출
df2 = df[df['연도'] == 2019]
```

```python
# '학제' 값들의 이름에서 () 삭제
'  (일반고)'.replace('  (','').replace(')','')
```

```python
# 전체 행에 적용할 함수 제작
def school_convert(x):
    try:
        return x.replace('  (','').replace(')','')
    except:
        return x
```

```python
school_convert(df2['학제'].iloc[4])
```

```python
df2['학제']=df2['학제'].apply(lambda x: school_convert(x))
```

```python
df2['학제'].unique()
```

```python
# '학제'에서 '고등학교' ,'일반고', '특목고', '특성화고' 만 빼고 삭제
df3 = df2[[i in ['고등학교' ,'일반고', '특목고', '특성화고']
          for i in df2['학제']]]
```

```python
df3 # 연도 2022, 학제(고등학교,일반고,특목고,특성화고) 만 저장된 데이터
```

```python
# 연도 칼럼 삭제
df3.drop(['연도'], axis=1,inplace=True)
```
