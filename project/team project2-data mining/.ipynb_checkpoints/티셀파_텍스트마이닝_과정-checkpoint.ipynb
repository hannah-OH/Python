```python
reviews = ""
```

```python
# 1000개의 리뷰 문장을 하나의 리뷰로 합치기 

for i in range(len(df_reviews)):
    reviews += df_reviews[i]
```

```python
def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
    return result
```

```python
# 한글과 띄어쓰기를 제외한 문자 삭제 적용

reviews = apply_regular_expression(reviews)
reviews
```

```python
# 명사 형태소 추출 
okt = Okt()
nouns = okt.nouns(apply_regular_expression(reviews))
nouns
```

```python
# 명사 형태소 개수 확인 

counter = Counter(nouns)
counter
```

```python
available_word = []
```

```python
# 한글자 명사 제거 

for i in nouns:
    if len(i) > 1:
        available_word.append(i)
```

```python
def countLetters(word):
    counter = pd.DataFrame(columns=('words', 'count'))
    
    for i in len(word):
        for letter in word:
            if letter not in counter:
                coun = 0
                counter[0, 'words'] = letter

            counter[i, 'count'] += 1
    return counter
```
