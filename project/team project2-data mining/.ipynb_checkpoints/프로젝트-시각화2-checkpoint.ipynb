```python
## 리뷰 데이터 시각화
```

```python
top10 = df_reviews.연수명.value_counts().sort_values(ascending=False)[:10]
top10_연수명 = top10.index.tolist()
top10_후기 = df_reviews[df_reviews['연수명'].isin(top10_연수명)]

print(top10_연수명)
print(top10_후기.info())
```

```python
# top10_연수명이 리스트인 경우, 이를 데이터프레임으로 변환
top10_연수명_df = pd.DataFrame({'연수명': top10_연수명})

# 중복을 제거하여 고유한 영화 제목 추출
연수명_이름 = top10_연수명_df['연수명'].unique().tolist()

# 평균 평점 계산을 위해 데이터프레임에서 해당 영화들의 평점을 추출하고 평균을 계산
avg_score = {}
for t in 연수명_이름:
    avg = top10_후기[top10_후기['연수명'] == t]['별점'].mean()
    avg_score[t] = avg

max_value = max(avg_score.values())  # 딕셔너리의 값들 중 최대값을 구함

for x, y in avg_score.items():
    color = np.array_str(np.where(y == max_value, 'orange', 'lightgrey'))    
    plt.bar(x, y, color=color)
    plt.text(x, y, '%.2f' % y, 
             horizontalalignment='center',  
             verticalalignment='bottom')
   

plt.show()
```

```python
pos_reviews = df_reviews[df_reviews['내용'] == 1]
neg_reviews = df_reviews[df_reviews['내용'] == 0]
```

```python
#-- 긍정 리뷰
pos_reviews['내용'] = pos_reviews['내용'].apply(lambda x: re.sub(r'[^ㄱ-ㅣ가-힝+]', ' ', x))
#-- 부정 리뷰
neg_reviews['내용'] = neg_reviews['내용'].apply(lambda x: re.sub(r'[^ㄱ-ㅣ가-힝+]', ' ', x))
```

```python
from konlpy.tag import Okt
okt = Okt()

pos_comment_nouns = []
for cmt in pos_reviews['내용']:
    pos_comment_nouns.extend(okt.nouns(cmt)) #-- 명사만 추출
#-- 추출된 명사 중에서 길이가 1보다 큰 단어만 추출
pos_comment_nouns2 = []
word = [w for w in pos_comment_nouns if len(w) > 1]  
pos_comment_nouns2.extend(word)
```

```python
pos_comment_nouns2
```

```python
from collections import Counter
pos_word_count = Counter(pos_comment_nouns2)

pos_word_count
```

```python
max = 20
pos_top_20 = {}
for word, counts in pos_word_count.most_common(max):
    pos_top_20[word] = counts
    print(f'{word} : {counts}')
```

```python
## 형태소 시각화
```

```python
plt.figure(figsize=(10, 5))
plt.title('긍정 리뷰의 단어 상위 (%d개)' % max, fontsize=17)    
plt.ylabel('단어의 빈도수')    
plt.xticks(rotation=70)
for key, value in pos_top_20.items():
    if key == '연수': continue
    plt.bar(key, value, color='lightgrey') 
plt.show()
```

```python
#-- 부정 리뷰 명사 추출
neg_comment_nouns = []
for cmt in neg_reviews['내용']:
    neg_comment_nouns.extend(okt.nouns(cmt)) 

neg_comment_nouns2 = []
word = [w for w in neg_comment_nouns if len(w) > 1]  
neg_comment_nouns2.extend(word)  
    
#-- 단어 빈도 계산
neg_word_count = Counter(neg_comment_nouns2)

#-- 빈도수가 많은 상위 20개 단어 추출
neg_top_20 = {}
for word, counts in neg_word_count.most_common(max):
    neg_top_20[word] = counts
    print(f'{word} : {counts}')

#-- 그래프 작성
plt.figure(figsize=(10, 5))
plt.title('부정 리뷰의 단어 상위 (%d개)' % max, fontsize=17)    
plt.ylabel('단어의 빈도수')    
plt.xticks(rotation=70)
for key, value in neg_top_20.items():
    if key == '연수': continue
    plt.bar(key, value, color='lightgrey') 
plt.show()
```
