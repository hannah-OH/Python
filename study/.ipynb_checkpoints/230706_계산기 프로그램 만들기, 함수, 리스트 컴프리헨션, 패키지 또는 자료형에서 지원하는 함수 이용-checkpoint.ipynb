## 계산기 프로그램 만들기 
1. 사친연산만 합니다.
2. 정수 두 개를 받습니다. 
3. 어떤 연산을 할지 물어봅니다. 
4. 선택한 연산에 따라 연산 함수를 실행합니다. 
5. 연산 함수의 결과값을 출력합니다. 

```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

def mul(x,y):
    return x * y

def div(x, y):
    if y !=0:
        return x / y
    
def calc():
    x = int(input('숫자 하나를 입력해주세요 : '))
    y = int(input('숫자를 하나 더 입력해주세요 : '))
    z = int(input('원하시는 연산의 번호를 입력해주세요. (1 덧셈 2 뺄셈 3 곱하기 4 나누기) : '))
    
    if z == 1 :
        return print("{} + {} = {}입니다." .format(x, y, add(x, y)))
    elif z == 2 : 
        return print("{} - {} = {}입니다." .format(x, y, sub(x, y)))
    elif z == 3 : 
        return print("{} x {} = {}입니다." .format(x, y, mul(x, y)))
    elif z == 4 : 
        if y != 0:
            return print("{} ÷ {} = {}입니다." .format(x, y, div(x, y)))
        else : 
            return print("0으로 나눌 수 없습니다.")
    else : 
        return print("잘못된 연산 번호입니다. 다시 입력해주세요.")
```

## 함수 실습 

연산을 하는 함수 만들기

[조건]
+ 함수 이름: func1
+ 입력받는 값: 임의의 숫자 둘(x, y)
+ 반환값: x / y의 나머지와 x - y를 비교해
+ 전자가 크면 Type I
+ 둘이 같거나 후자가 크면 Type II라고 반환

```python
def finc1(x, y):
    
    z1 = x % y
    z2 = x - y
    
    if z1 > z2:
        return 'Type I'
    else :
        return 'Type II'
```

```python
finc1(9, 8)
```

### 리스트 컴프리헨션
+ 장점 : 간편함, 속도가 항상 빠름
+ 단점 : 가지수가 늘어나면 안됨, 순서가 복잡(어렵)

```python
# 순서 복잡

[i for i in range(10) if i%2 == 0]
```

```python
# 함수를 이용해서 더 간편하게

def funcY(x):
    if x%2 == 0 :
        return x
    else : 
        return 0
```

```python
[funcY(i) for i in range(10)]
```

```python
# 식이 너무 길다면? 함수 이용

def funcZ(i):
    return i+1-4*2-3

[funcZ(i) for i in range(10)]
```

##### 리스트 컴프리헨션과 apply 공통점 
+ for문보다 빠르다 
+ 리스트로 리턴 가능

##### apply를 쓰는 이유?
+ 기하급수적으로 늘어날때 더 빠를 수 있음
+ 사람 논리로 더 가깝기때문에 > 유지보수가 쉬움

+ df 전체 적용이 가능함 

##### 어찌되었든, 리스트 컴프리헨션, apply, for문 다 읽을 줄 알아야하며, 하나 이상은 자유자재로 쓸 줄 알아야한다.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('한국전력거래소_시간별 제주 풍력 발전 제어량_20211231.csv', encoding='cp949')
```

```python
df.head(2)
```

```python
# 1. 일자 정보를 정리한다: YYYY-MM-DD -> DD 형식으로 바꾸기. 단, 해당 날짜는 계산이 가능해야한다.

df['일자'] = [int(i.split('-')[-1]) for i in df['일자']]
```

```python
df
```

```python
# 모범답안

def date(x):
    split_date = x.split('-')
    split_date = split_date[-1]
    return int(split_date)
```

```python
[date(i) for i in df['일자']]
```

```python
df['일자'].apply(lambda x : date(x))
```

```python
df['일자']
```

```python
# 2. 모든 일자에 대해 1~24시 정보 중, 몇 시에 제어량이 가장 높은지 알아보자.

df.loc[:, '1시':'24시'].max(axis=1)

# max값은 구했는데, 일자에 대한 몇시인지를 어떻게 출력할지 고민중 
```

```python
# 모범답안

for i in df.index:
    print(df.loc[i, '1시':'24시'].max())
```

```python
df['일자별 제어량 최대값'] = [df.loc[i, '1시':'24시'].max() for i in df.index]
```

```python
df.head(3)
```

```python
df.apply(lambda x: x) # x: x 는 데이터프레임 전체 대상을 의미함 

df.apply(lambda x: x['1시':'24시']) 

# apply 후 인덱싱은 불가능
# 그러므로 인덱싱 먼저하고 apply를 해라.
```

```python
# 3. 총제어량이 가장 낮은 날짜를 구하라.

data = 1000

for i in range(len(df)):
    if data > df['총제어량'][i]:
        data = df['총제어량'][i]
        data_min = df['일자'][i]
        
print("일자는 {}일이고, 총 제어량은 {}입니다." .format(data_min, data))
```

```python
# 다른방법 : 마스킹 
# 마스킹 : True 혹은 False 기반의 인덱싱

df[df['총제어량'] == 0.84]
```

```python
df['총제어량'].mean()
```

```python
df['평균'] = 0
```

```python
# 4. 총제어량이 평균 이상인 대상을 True, 아니면 False로 표기해서 평균 column으로 저장한다.

mean = df['총제어량'].mean()

for i in range(len(df)):
    if df['총제어량'][i] > mean:
        df['평균'][i] = "True"
    else : 
        df['평균'][i] = "False"
```
