```python
# 에러발생 원인 : len~ 순서기반으로 하고 있었는데, 89번 데이터가 존재하지 않아서 에러.
# 이럴땐 숫자 기반으로 인덱싱해야한다. >> iloc ! 숫자만 입력해야함을 주의하자

df1['짝수'] = 0

for i in range(len(df1)):
    if df1.iloc[i, 13] % 2 == 0:
        df1['짝수'][i] = 'True'
    else :
        df1['짝수'][i] = 'False'
```

```python
# 모범답안

def to_int(x):
    try:
        x = int(x)
        
        if x % 2 == 0 :
            return True
        else :
            return False
        
    except:
        return False
    
# try : 일단 실행함. 만약 에러가 발생하면 바로 except로 넘어감. 에러가 발생하지 않으면 PASS
```

```python
m1 = df['강좌정원수'].apply(lambda x: to_int(x))
```

```python
df[m1]
```

```python
df = pd.read_csv('./시청자미디어재단_체험형미디어교육_20230704.csv', encoding='cp949')
```

```python
def to_int2(x):
    try:
        return int(x)
    except:
        return 0
```

```python
df['강좌정원수2'] = df['강좌정원수'].apply(lambda x: to_int2(x))
```

```python
df[df['강좌정원수2'] < 10]
```

```python
# df[0 < df['강좌정원수2'] < 10] 는 안되지만, 쿼리를 사용하면 된다. 

query_expr = "0 < 강좌정원수2 < 10"
df.query(query_expr)
```
