### 결측치를 왜 쓰는지 ? 
#### 결측치 대신 0, '', ' ', 파이 기호 를 쓰지않고 결측치를 쓰는 이유 ?

[이유]
+ ' ', ''가 들어있으면 연산시 오류가 발생되므로 사용하지 않음.
+ 0은 무엇을 곱해도 0이 되기때문에 다른 데이터를 없애버릴 위험이 있으므로 사용하지 않음. 

### 그럼 nan 값은 이런 문제가 전혀 없느냐? 그렇다.
+ 덧셈 뺄셈 나누기 곱하기 순서를 바꿔도 그냥 nan값으로 처리됨 
+ 문제가 되는 연산에 대해서는 nan이 존재한다는 것을 바로 알 수 있기 때문에 (0은 정상값이 0인지 잘못된 값이 0인지 판단하기 어려움 )

### nan값의 특이한 특징
+ 자기 자신과 일치하지 않음 
+ 그러므로 결측치가 있는지 없는지 판단하기 편하다

    - 장점 : 속도가 빠름 / 이상한 데이터가 대표한다고 착각할 확률이 줄어듦 / 비용이 더 적음 
- 클러스터링
    - 비슷한 것 끼리 묶는거 
- 데이터 용량 축소 
    - 데이터를 정보로 바꾸는 전처리 과정에 엄밀히 말하면 들어가진 않지만, 그냥 과정에서 병행되는 편

 - 연속형은 기준이 없다. 컴퓨터는 판단을 할 수가 없다. 그것은 데이터 이산화를 통해 기준을 만들어주면 해결된다.

```python
# 해설

target = ['일반시민', '초등생이상', '시니어', '초등생 이상', '일반', '일반 시민']
```

```python
m = media['교육대상구분'].apply(lambda x: x in target)
```

```python
df2 = media[m]
```

```python
def to_int(x):
    try:
        return int(x)
    except:
        return 0
    
# 지금까진 계속 int로 하고 있긴하지만, 데이터사이언티스트 입장에선 float로 받는것이 맞다. 
# 그렇긴한데 ... 정안될때만 실수를 하자 
# 그리고 정수는 절대 실수로 바꾸지 않도록 하자 
# 왜냐하면, 파이썬은 소수점을 다룰때 부동소수점을 사용함. (치명적인 문제)
# 파이썬은 CPU, GPU가 있는데 서로 다름. CPU는 정확해야하고 GPU는 살짝정도 틀어진거정돈 괜찮아하기 때문
```

```python
# 고정 소수점
# 소수점 자리의 정확한 자리수를 측정해서 정확히 알려줌

# 부동 소수점
# 컴퓨터가 근사값 몇개를 준비해서 비슷한거 있으면 그거 준다. -> 속도는 더 빠르기 때문
```

```python
# 딥러닝 문제는 0부터 1까지 정규화하니까 소수점이 큰게 나오는데 부동 소수점이 사용된다.
# 그러므로 이거다 아니다 딱 말할 수 없다 
# 그래서 결과값을 &&일 확률은 @@%다. 이런식으로 말하는 것이다. 
```

```python
df2['강좌정원수'] = df2['강좌정원수'].apply(lambda x: to_int(x))

# 이 경우는 얕은 복사를 의도하고 한 것. 
# 얕은 복사를 지원하지않았더라면 이런 건 할 수 없었을거임.
```
